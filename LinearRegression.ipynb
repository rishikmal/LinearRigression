## Creating New Dummy Variables

```python
## Inspect the Data Frame first to determine Categorical Variables Manually
cars.head(10)
```

```python
### Splitting Company Name from CarName Variable
cars['companyName'] = cars.CarName.str.split(' ').str.get(0)
cars['companyName'].value_counts()
```

```python
## Correcting the Spelling mistake on Porche, Volkwagen 
cars.companyName = cars.companyName.str.replace("porcshce", "porsche")
cars.companyName = cars.companyName.str.replace("vokswagen", "volkswagen")
cars.companyName = cars.companyName.str.replace("vw", "volkswagen")
cars['companyName'].value_counts()
```

### Print Unique values and their count for Categorical Variables

```python
## Printing out Unique values in Categorical Variables
print(cars.symboling.value_counts())
print(cars.cylindernumber.value_counts())
print(cars.doornumber.value_counts())
print(cars.enginetype.value_counts())
print(cars.enginelocation.value_counts())
print(cars.aspiration.value_counts())
print(cars.carbody.value_counts())
print(cars.drivewheel.value_counts())
print(cars.fuelsystem.value_counts())
print(cars.fueltype.value_counts())
```

```python
# Create a new Datafrane with Dummies for Categorical Variables except Symboling which is already good to go
newStatuses = pd.get_dummies(cars[['doornumber', 'cylindernumber', 'drivewheel', 'aspiration', 'carbody', 'fuelsystem',
                                   'fueltype', 'enginelocation', 'enginetype', 'companyName']], drop_first = True)
print(newStatuses.info())
newStatuses.head(25)
```

```python
## Concatinate with the Original Dataframe to get the final Dataframe
cars = pd.concat([cars, newStatuses], axis=1)
cars.head()
```

```python
## Check Shape and column dtypes of new DF
print(cars.shape)
cars.info()
```

```python
## Dropping Variables converted to Dummy Variables.
cars = cars.drop(['doornumber', 'cylindernumber', 'drivewheel', 'aspiration', 'carbody', 'fuelsystem', 
                  'fueltype','enginelocation', 'enginetype', 'companyName'], axis=1)
```

### Creating a new DataFrame with all numerical Variables

```python
## Checking the Column Dtypes and Shape of the final DF
cars.info()
cars.shape
pd.set_option('display.max_columns', 75)
```

```python
## Selecting only Numerical Columns from DF.
carsAllNum = cars.select_dtypes(include=['float64', 'int64', 'int8', 'uint8'], exclude='object')
carsAllNum.head()
```

```python
##We can drop the car_ID field as it is just an index for cars.
carsAllNum = carsAllNum.drop(['car_ID'], axis=1)
carsAllNum.head()
```

### Splitting Dataframe into Test and Train Dataset
Now that we have got our new Dataframe in perfect shape we should split the dataframe into test and train Dataset

```python
#Seeding to make sure we have the same rows.
np.random.seed(0)

## Splitting the Dataframe into 70% Train and 30% Test Data
from sklearn.model_selection import train_test_split
cars_train, cars_test = train_test_split(carsAllNum, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
print(cars_train.shape)
cars_train.head()
```

```python
# Let's check the correlation coefficients to see which variables are highly correlated

plt.figure(figsize = (40, 30))
sns.heatmap(cars_train.corr(), cmap="YlGnBu")
plt.show()
```

### Scalling Variables to bring them all within the same scale

```python
## Using MinMaxScaler from sklearn 
from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler()
```

```python
## Finding out Variables for Scalling, Variables except the Dummy and Boolean Variables are chosen.
DummyVars = ['wheelbase', 'carlength', 'carwidth', 'carheight', 'curbweight', 'enginesize', 'boreratio',
            'stroke', 'compressionratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg', 'price']
```

```python
## Scaling Variables to bring them on a same level
cars_train[DummyVars] = scale.fit_transform(cars_train[DummyVars])
cars_train.head()
```

```python
##Creating Variables for X and Y used for Model Building
y_train = cars_train.pop('price')
x_train = cars_train
```

## Building Model

### Using RFE(Recursive Feature Elimination) for Feature Selection

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
## Running RFE and keeping the Number of Variables to 20
lr = LinearRegression()
lr.fit(x_train, y_train)

rfe = RFE(lr, 10)
rfe = rfe.fit(x_train, y_train)
```

```python
## Print Variables with thier output for RFE and ranking
list(zip(x_train.columns, rfe.support_, rfe.ranking_))
```

```python
## Variables Selected by RFE which influences values of the target variable price
col = x_train.columns[rfe.support_]
col
```

```python
## Print Variables discarded by RFE
x_train.columns[~rfe.support_]
```

### Building a Stats Model

```python
## with selected RFE models creating a new DF
x_trainRfe = x_train[col]
```

```python
## Adding a constant Variable
import statsmodels.api as sm
x_trainRfe = sm.add_constant(x_trainRfe)

## Copying originally Created DF to another DF
x_trainRfeOrg = x_trainRfe
```

```python
#Building a Model using OLS
statsM = sm.OLS(y_train, x_trainRfe).fit()
```

```python
statsM.summary()
```

```python
## Dropping Variables with high p Value i.e. cylindernumber_twelve.
x_trainRfe = x_trainRfe.drop(["cylindernumber_twelve"], axis=1)
```

```python
## Rebuilding the Model
import statsmodels.api as sm
statsM = sm.OLS(y_train, x_trainRfe).fit()
statsM.summary()
```

```python
## p Values for all the variable are will under .05. So calculate the VIF for these variables.
x_trainRfe = x_trainRfe.drop("const", axis=1)
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
x = x_trainRfe
vif["Features"] = x.columns
vif["VIF"] = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]
vif["VIF"] = round(vif["VIF"], 2)
vif = vif.sort_values(by = "VIF", ascending=False)
vif
```

```python
## There are many Variable with high VIF Values so dropping them one by one and checking statistics with VIF
## Drop curbweight as it has the hieghest VIF Score of 38.41
x_trainRfe = x_trainRfe.drop(["curbweight"], axis=1)

## and rebuilding the Model again
import statsmodels.api as sm
x_trainRfe = sm.add_constant(x_trainRfe)
statsM = sm.OLS(y_train, x_trainRfe).fit()
print(statsM.summary())

x_trainRfe = x_trainRfe.drop("const", axis=1)
## Checking the VIF score 
vif = pd.DataFrame()
x = x_trainRfe
vif["Features"] = x.columns
vif["VIF"] = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]
vif["VIF"] = round(vif["VIF"], 2)
vif = vif.sort_values(by = "VIF", ascending=False)
vif
```

```python
## Now the p Values are spiked a little as well but cannot trust the p Value as high VIF Values suggest Multicollinearity
## so dropping variables with high VIF Score and checking statistics with VIF. Next is carwidth with VIF Score 16.48
x_trainRfe = x_trainRfe.drop(["carwidth"], axis=1)

## and rebuilding the Model again
import statsmodels.api as sm
x_trainRfe = sm.add_constant(x_trainRfe)
statsM = sm.OLS(y_train, x_trainRfe).fit()
print(statsM.summary())

x_trainRfe = x_trainRfe.drop("const", axis=1)
## Checking the VIF score 
vif = pd.DataFrame()
x = x_trainRfe
vif["Features"] = x.columns
vif["VIF"] = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]
vif["VIF"] = round(vif["VIF"], 2)
vif = vif.sort_values(by = "VIF", ascending=False)
vif
```

```python
## Now that the boreratio is 5.49 which is still considerable so lets eliminate variables based on p values now
x_trainRfe = x_trainRfe.drop(["boreratio"], axis=1)

## and rebuilding the Model again
import statsmodels.api as sm
x_trainRfe = sm.add_constant(x_trainRfe)
statsM = sm.OLS(y_train, x_trainRfe).fit()
print(statsM.summary())

x_trainRfe = x_trainRfe.drop("const", axis=1)
## Checking the VIF score 
vif = pd.DataFrame()
x = x_trainRfe
vif["Features"] = x.columns
vif["VIF"] = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]
vif["VIF"] = round(vif["VIF"], 2)
vif = vif.sort_values(by = "VIF", ascending=False)
vif
```

```python
## Now we have got all the VIF Values under 5 and considering p Values for eliminating Variables.
#cylindernumber_three has the hieghest p value so eliminating the cylindernumber_three variable
x_trainRfe = x_trainRfe.drop(["cylindernumber_three"], axis=1)

## and rebuilding the Model again
import statsmodels.api as sm
x_trainRfe = sm.add_constant(x_trainRfe)
statsM = sm.OLS(y_train, x_trainRfe).fit()
print(statsM.summary())

x_trainRfe = x_trainRfe.drop("const", axis=1)
## Checking the VIF score 
vif = pd.DataFrame()
x = x_trainRfe
vif["Features"] = x.columns
vif["VIF"] = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]
vif["VIF"] = round(vif["VIF"], 2)
vif = vif.sort_values(by = "VIF", ascending=False)
vif
```

```python
## Next is enginelocation_rear with p value .226
#x_traninRfeOr1 = x_trainRfe
x_trainRfe = x_trainRfe.drop(["enginelocation_rear"], axis=1)

## and rebuilding the Model again
import statsmodels.api as sm
x_trainRfe = sm.add_constant(x_trainRfe)
statsM = sm.OLS(y_train, x_trainRfe).fit()
print(statsM.summary())

## Hoping to get the Final Set of Variables
x_trainRfeFinal = x_trainRfe

x_trainRfe = x_trainRfe.drop("const", axis=1)
## Checking the VIF score 
vif = pd.DataFrame()
x = x_trainRfe
vif["Features"] = x.columns
vif["VIF"] = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]
vif["VIF"] = round(vif["VIF"], 2)
vif = vif.sort_values(by = "VIF", ascending=False)
vif
```

### We have Finally got p Values and VIF Score well under control so we can move ahead with Residual Analysis
